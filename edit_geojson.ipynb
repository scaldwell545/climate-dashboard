```python
## Import Dependencies
import os
import pandas as pd
import csv
import json
import numpy as np
```

```python
temp_change_path = os.path.join(os.getcwd(), 'Resources', 'Environment_Temperature_change_E_All_Data_NOFLAG.csv')
```

```python
temp_change_df = pd.read_csv(temp_change_path, encoding='latin-1')
```

```python
temp_change_df.head()
```

```python
temp_change_df.loc[temp_change_df.Area == "Russian Federation", 'Area'] = 'Russia'
```

```python
## Create a df with yearly changes
filtered_df = temp_change_df[(temp_change_df["Element"] == "Temperature change") & (temp_change_df["Months"] == "Meteorological year")]

## Drop unrelated data
cleaned_df = filtered_df.drop(["Area Code", "Months Code", "Months", "Element", "Element Code", "Unit"], axis=1)
 
## Transpose
transposed_df = cleaned_df.T
transposed_df.columns = transposed_df.loc["Area"]
transposed_df.drop("Area", inplace=True)
transposed_df.to_csv("Resources/yearly_change.csv")
transposed_df.head()
```

```python
transposed_df = transposed_df.reset_index()
transposed_df["Year"] = transposed_df["index"].str.replace("Y", "")
transposed_df.head()
```

```python
transposed_df = transposed_df.set_index('Year').drop("index", 1)
transposed_df.head()
```

```python
array_of_tuples = []
for row in transposed_df.itertuples():
    i = 1
    for column in transposed_df:
        new_array = [getattr(row, 'Index'), column, row[i]]
        array_of_tuples.append(new_array)
        i = i+1
```

```python
array_of_tuples[0]
```

```python
df = pd.DataFrame(array_of_tuples, columns=['Year', 'Country', 'Value']).fillna('')
d = {}
for i in df['Country'].unique():
    d[i] = {df['Year'][j]: df['Value'][j] for j in df[df['Country']==i].index}
```

```python
for key, value in d.items():
    print(key)
    #print(value)
```

```python
# Opening JSON file
f = open('Resources/geo_info/archive/countries.geojson',)
 
# returns JSON object as
# a dictionary
data = json.load(f)
 
# Iterating through the json
# list

for i in data['features']:
    for key, value in d.items():
        if i['properties']['ADMIN'] == key:
            i['properties'].update({ "temp" : value })         
            
## Save file            
with open('Resources/newcountry.geojson', 'w') as fp:
    json.dump(data, fp)
```
